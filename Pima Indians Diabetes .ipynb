```python
import numpy as np
import pandas as pd
```

####Reading the Dataset

```python
df=pd.read_csv('diabetes.csv')
df.head()
```

```python
df.info()
```

```python
df.corr()
```

```python
df.drop(['SkinThickness'],axis=1,inplace=True)
```

```python
df
```

```python
df.skew()
```

```python
df.DiabetesPedigreeFunction=np.log(df['DiabetesPedigreeFunction'])
```

```python
df.Age=np.log(df['Age'])
```

```python
df.BloodPressure=np.sqrt(df['BloodPressure'])
```

```python
df.BloodPressure=np.sqrt(df['BloodPressure'])
```

```python
df.Insulin=np.sqrt(df['Insulin'])
```

```python
df.skew()
```

####Designing a Model

```python
y=df.Outcome
col=(['Pregnancies','Glucose','Insulin','BMI','DiabetesPedigreeFunction','Age','BloodPressure'])
x=df[col]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
model.fit(X_train,y_train)
```

```python
y_prediction=model.predict(x_test)
```

```python
y_prediction
```

```python
model.score(x_test,y_test)
```

###Deploying the Model
